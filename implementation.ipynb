```python
smoking_map = {
    'never':0,
    'No Info':1,
    'current':2,
    'former':3,
    'ever':2,
    'not current':-1}
```

```python
df['smoking_history'] = df['smoking_history'].map(smoking_map)
```

```python
x = df[['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history',
       'bmi', 'HbA1c_level', 'blood_glucose_level']]
```

```python
y = df[['diabetes']]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.10,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cls = LogisticRegression()
```

```python
cls.fit(x_train , y_train)
```

```python
result = cls.predict(sc.transform([[0,54.00,0,0,-1,27.32,6.60,80.00]]))
```

```python
result = result[0]
```

```python
if result == 0:
    print("Person is not diabetic")
else:
    print("person is diabetic")
```

```python
y_pred = cls.predict(x_test)
```

```python
y_pred = pd.DataFrame(data=y_pred,columns=['result'])
```

```python
result = pd.concat((y_test.reset_index(drop=True),y_pred.reset_index(drop=True)),axis=1)
```

```python
result
```

```python
from  sklearn.metrics import accuracy_score,confusion_matrix
```

```python
cm = confusion_matrix(result['diabetes'],result['result'])
cm
```

```python
score = accuracy_score(result['diabetes'],result['result'])
score
```
